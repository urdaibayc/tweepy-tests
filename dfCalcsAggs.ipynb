```python
import pandas as pd
import numpy as np
pd.set_option("display.float_format", lambda x: "%.5f" % x)
pd.set_option("display.max_columns", 20)
pd.set_option("display.min_rows", 50)
pd.set_option("display.max_rows", 50)
```

```python
data = pd.read_csv("supermarket_sales.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
#data["Date"] = data['Date'].astype("datetime64[ns]")
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
data['Days'] = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
data['Q'] = data['Date'].dt.to_period("Q")
```

```python
data.head()
```

```python
data['Time'] = pd.to_datetime(data['Time'])
data['Hour'] = data['Time'].dt.hour
```

```python
data.head()
```

```python
data['Minute'] = data['Time'].dt.minute
```

```python
data.head()
```

```python
data['Hour'].nunique()
```

```python
np.sort(data['Hour'].unique())
```

```python
data.describe()
```

```python
menorAlPercentil50 = data['Unit price'] < data.describe().loc["50%", "Unit price"]
data[menorAlPercentil50]['Unit price'].count()
```

```python
categorical_columns = [c for c in data.columns if data[c].dtype == "object"]
categorical_columns
```

```python
branches = (
    data
      .groupby("Branch")
)
branches.groups
```

```python
branches = (
    data
      .groupby("Branch")
      .count()
)
branches
```

```python
branches = (
    data
      .groupby("Branch")
      .sum()
)
branches
```

```python
branches = (
    data
      .groupby("Branch")
      .mean()
)
branches
```

```python
branches = (
    data
      .groupby("Branch")
      .min()
)
branches
```

```python
branches = (
    data
      .groupby("Branch")
      .max()
)
branches
```

```python
productLine = (
    data
      .groupby(["Branch", "Product line"])
      [["Unit price", "Total"]]
      .sum()
)
productLine
```

```python
productLine['Rank'] = productLine['Total'].rank(ascending=False)
productLine['Rank_UnitPrice'] = productLine['Unit price'].rank(ascending=True)
productLine.sort_values("Rank")
```

```python
cityGender = (
    data
      .groupby(["City", "Gender"])
      .agg({
          "Quantity" : np.sum,
          "Total": np.mean
      })
      .sort_values("Gender")
)
cityGender
```

```python
data.columns
```

```python
genrePayment = (
    data
      .groupby(["Gender", "Payment"])
      .agg(
          total_invoice = ("Invoice ID", "count"),
          client_satisfaction = ("Rating", "mean")
      )
)
genrePayment
```

```python
pivote = (
    pd
      .pivot_table(
          data,
          values=['Total'],
          columns=['Hour'],
          index=['Product line'],
          aggfunc={"Total": np.sum}
      )
)
pivote
```

```python
pivote.to_excel("reporteParaeljefelatoso.xlsx")
```

```python
pivote.sort_values(by=list(pivote.columns), axis=0, ascending=False)
```

```python
pivote = (
    pd
      .pivot_table(
          data,
          values=['Total'],
          columns=['Hour'],
          index=['Product line', "City"],
          aggfunc={"Total": np.sum}
      )
)
pivote
```

```python
data.columns
```

```python
pivote = (
    pd
      .pivot_table(
          data,
          values=['Total', "Quantity"],
          columns=['Hour'],
          index=['Product line', "City"],
          aggfunc={"Total": np.sum, "Quantity": "count"}
      )
)
pivote
```

```python
pivote.T
```

```python
pivote.unstack(level=-2, fill_value=0)
```

```python
pivote.stack()
```

```python
crosstab = (
    pd.crosstab(
        data['Gender'],
        data['Customer type']
    )
)
crosstab
```

```python
crosstab = (
    pd.crosstab(
        data['Gender'],
        data['Customer type'],
        margins=True
    )
)
crosstab
```

```python
crosstab = (
    pd.crosstab(
        data['Gender'],
        data['Customer type'],
        margins=True,
        normalize= True
    )
)
crosstab
```

```python
crosstab = (
    pd.crosstab(
        data['Gender'],
        data['Customer type'],
        values = data['Tax 5%'],
        aggfunc="sum",
        margins=True,
        normalize=True
    )
)
crosstab
```
